```python
campaigns =[]
```

```python
#campaigns with word republican
driver=webdriver.Chrome()
driver.get('https://www.gofundme.com/mvc.php?route=homepage_norma/search&term=republican')
show_me_count = 0
button = True
while button:
    wait = WebDriverWait(driver, 10)
    element = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR,'body > div.content-section.content-section--tan > div > div > div.cell.small-12.text-center.mt2x > a'))) 
    button = driver.find_element_by_css_selector('body > div.content-section.content-section--tan > div > div > div.cell.small-12.text-center.mt2x > a')
    button.click()
    show_me_count +=1
    print(show_me_count)
```

```python
#putting in BeautifulSoup for scraping 
html = driver.page_source
doc = BeautifulSoup(html)
```

```python
campaign_tiles = doc.find_all('div', class_='react-campaign-tile-details')
for campaign in campaign_tiles:
    tile={}
    tile['location'] =campaign.find('span').text
    link = campaign.find('a')
    tile['title']=link.find('div',class_='fund-title clamp hide-for-medium').text
    tile['link'] =link['href']
    funds = campaign.find(class_='fund-item truncate-single-line')
    tile['funds']=funds.find(class_='show-for-medium').text
    summary = campaign.find(class_='fund-item fund-description clamp mb')
    try:
        tile['summary']=summary.text.strip()
    except:
        tile['summary']='summary was NoneType'
    campaigns.append(tile)
```
