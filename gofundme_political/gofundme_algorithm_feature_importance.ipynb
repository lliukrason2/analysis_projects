```python
pd.set_option('display.max_columns', 500)
```

```python
df=pd.read_csv('cleaned_08072019.csv')
```

```python
df.info()
```

```python
df['is_dem'] = np.where(df['tag'] == 'dem', 1,0)
```

```python
count_vectorizer = CountVectorizer()
X = count_vectorizer.fit_transform(df.summary)
X
```

```python
words_df = pd.DataFrame(X.toarray(), columns=count_vectorizer.get_feature_names())
```

```python
words_df
```

```python
y=df['is_dem']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
clf = RandomForestClassifier()

model = RandomForestClassifier(n_estimators = 10, max_depth=5)
model.fit(X_train, y_train)
```

```python
feature_names = words_df.columns
importances = model.feature_importances_

pd.DataFrame({
    'feature': feature_names,
    'feature importance': importances,
}).sort_values(by='feature importance', ascending=False)
```

```python
y_true = y_train
y_pred = model.predict(X_train)
matrix = confusion_matrix(y_true, y_pred)

label_names = pd.Series(['not dem', 'dem'])
pd.DataFrame(matrix,
     columns='Predicted ' + label_names,
     index='Is ' + label_names)
```
