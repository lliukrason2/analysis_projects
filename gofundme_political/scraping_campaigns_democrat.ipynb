```python
import pandas as pd
import requests
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import re
import time
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions as EC

%matplotlib inline
```

```python
campaigns =[]
```

```python
#network--> XHR --> request URL
#apparently, you can paginate based on the campaigns
#there are 1325 results and 9 per page so 147 pages
pages= range(1,147)
```

```python
#grabbing campaign, link, amount raised.
for page in pages:
    url = f'https://www.gofundme.com/mvc.php?route=homepage_norma/load_more&page={page}&term=democrat&country=&postalCode=&locationText='
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    campaign_tiles = doc.find_all('div', class_='react-campaign-tile-details')
    print("PAGE",page)
    for campaign in campaign_tiles:
        tile={}
        tile['location'] =campaign.find('span').text
        link = campaign.find('a')
        print(link.find('div',class_='fund-title clamp hide-for-medium').text)
        tile['title']=link.find('div',class_='fund-title clamp hide-for-medium').text
        tile['link'] =link['href']
        funds = campaign.find(class_='fund-item truncate-single-line')
        tile['funds']=funds.find(class_='show-for-medium').text
        campaigns.append(tile)
    print('---------------------------------------------')
```

```python
created =[]
failed =[]
```

# Grabbing Dates

```python
for link in dem_campaigns['link']:
    page={}
    print(link)
    page['link']=link
    response = requests.get(link)
    doc = BeautifulSoup(response.text)
    date=doc.find(class_='created-date')
    summary=doc.find(class_='o-campaign-story')
    
    try:
        page['created_date']=date.text
        
    except:
        try:
            page['created_date']=doc.find(class_='m-campaign-byline-created').text
        except:
            print(f"{link} didn't have.text") 
            failed.append(page)
            continue
    if summary!= None:
            page['summary']=summary.text
    created.append(page)
```

```python
dem_created=pd.DataFrame(created)
```

```python
dem_merge = dem_campaigns.merge(dem_created,left_on='link',right_on='link',how='left')
```
