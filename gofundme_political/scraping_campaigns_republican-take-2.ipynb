```python
import pandas as pd
import requests
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import re
import time
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions as EC

%matplotlib inline
```

```python
campaigns =[]
```

```python
#gofundme has 2551 campaigns with the word 'republican'
pages= range(1,284)
```

```python
#grabbing campaign, link, amount raised.

#grabbing campaign, link, amount raised.
for page in pages:
    url = f'https://www.gofundme.com/mvc.php?route=homepage_norma/load_more&page={page}&term=republican&country=&postalCode=&locationText='
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    campaign_tiles = doc.find_all('div', class_='react-campaign-tile-details')
    print("PAGE",page)
    for campaign in campaign_tiles:
        tile={}
        tile['location'] =campaign.find('span').text
        link = campaign.find('a')
        print(link.find('div',class_='fund-title clamp hide-for-medium').text)
        tile['title']=link.find('div',class_='fund-title clamp hide-for-medium').text
        tile['link'] =link['href']
        funds = campaign.find(class_='fund-item truncate-single-line')
        tile['funds']=funds.find(class_='show-for-medium').text
        campaigns.append(tile)
    print('---------------------------------------------')
```

```python
created =[]
failed =[]
```

```python
rep_campaigns.tail(10)
```

```python
for link in rep_campaigns['link']:
    page={}
    print(link)
    page['link']=link
    response = requests.get(link)
    doc = BeautifulSoup(response.text)
    date=doc.find(class_='created-date')
    summary=doc.find(class_='o-campaign-story')
    
    try:
        page['created_date']=date.text
        
    except:
        try:
            page['created_date']=doc.find(class_='m-campaign-byline-created').text
        except:
            print(f"{link} didn't have.text") 
            failed.append(page)
            continue
    if summary!= None:
            page['summary']=summary.text
    created.append(page)
```

```python
rep_created=pd.DataFrame(created)
```

```python
rep_merge = rep_campaigns.merge(rep_created,left_on='link',right_on='link',how='left')
```
